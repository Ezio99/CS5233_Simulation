```python
# analyze rate of arrival in 
visits_df = pd.read_csv("Data/visits.csv")
visits_df['A_to_B'] = 2 / visits_df['speed'] * 60
visits_df['B_to_C'] = 2.5 / visits_df['speed'] * 60
visits_df['C_to_D'] = 3 / visits_df['speed'] * 60
visits_df['D_to_A'] = 1 / visits_df['speed'] * 60
```

```python
stations = ['A', 'B', 'C', 'D']
arrival_rate = [[],[],[],[]]
dist = [1,2,2.5, 3]
for i, row in visits_df.iterrows():
    visit = row['visits']
    source = row['source']
    indexes = []
    for i in range(1, visit):
        next = stations[(visit+i) % len(stations)]
        arrival_rate[(visit+i-1) % len(stations)].append(dist[(visit+i) % len(stations)] / row['speed']* 60) 
        # print(stations[(visit+i-1) % len(stations)], next, dist[(visit+i) % len(stations)])

for rates, text_name in zip(arrival_rate, ["D_to_A", "A_to_B", "B_to_C", "C_to_D"]):
    with open(f"Data/{text_name}.txt", "w") as f:
        for r in rates:
            f.write(f"{r}\n")
```
