```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
folder_path = "./Data"
```

```python
df = pd.read_csv(os.path.join(folder_path, "visits.csv"))
df
```

```python
visits_df = df.drop(columns=["Unnamed: 2"])
```

```python
visit_counts = visits_df["visits"].value_counts().sort_index()
visit_counts
```

```python
plt.figure(figsize=(10, 6))
plt.bar(visit_counts.index, visit_counts.values, width=0.6)
plt.xlabel("Number of Attractions Visited")
plt.ylabel("Number of Tourists")
plt.title("Distribution of Tourists by Number of Attractions Visited")
plt.xticks(visit_counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
